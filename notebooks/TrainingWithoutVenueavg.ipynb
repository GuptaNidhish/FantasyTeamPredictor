```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/Users/nidhishgupta/Desktop/Dream11_Fantasy Team_Predictor/data/processed/ready_for_training.csv")
```

```python
pd.set_option("display.max_columns",None)
```

```python
data
```

```python
data['recent_form'] = (data['last3_avg_points']*0.6 + data['last5_avg_points']*0.3 + data['last10_avg_points']*0.1)
```

```python
data['venue_form'] = data['venue_avg_points'] * data['recent_form']
```

```python
colsToDrop = ['player_match_number',
              'last5_avg_points',
              'last10_avg_points',
              'bat_pos_per_match',
              'match_month',
              'stage_Elimination Final',
              'stage_Eliminator',
              'stage_Final',
              'stage_Qualifier 1',
              'stage_Qualifier 2',
              'stage_Semi Final',
              'stage_Unknown','venue_avg_points']
data = data.drop(columns = colsToDrop)
```

```python
data['season'] = data['season'].str[:4].astype(int)
data['season'].unique()
```

```python
train = data[data['season'] <= 2022]
test  = data[data['season'] > 2022]
X_train = train.drop(columns=['fantasy_points','player','venue','season','player_role_batsman','player_role_bowler'])
y_train = train['fantasy_points']

X_test = test.drop(columns=['fantasy_points','player','venue','season','player_role_batsman','player_role_bowler'])
y_test = test['fantasy_points']
print(X_test.shape)
print(X_train.shape)
```

```python
from xgboost import XGBRegressor

xg = XGBRegressor(
    n_estimators=900,
    learning_rate=0.01,
    max_depth=4,
    subsample=0.85,
    colsample_bytree=0.8,
    min_child_weight=3,
    gamma=0.1,
    random_state=42
)

xg.fit(X_train, y_train)
```

```python
y_pred = xg.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test, y_pred))
```
