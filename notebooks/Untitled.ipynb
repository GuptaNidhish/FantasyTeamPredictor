```python
data['venue'] = data['venue'].replace(venue_map)
```

```python
data.isnull().sum()
```

```python
data = data.drop(columns = ['venue_avg_points','venue_run_factor'])

match_runs = data.groupby("match_id")["runs"].sum().reset_index()
match_runs.rename(columns={"runs": "total_match_runs"}, inplace=True)
match_runs = match_runs.merge(
    data[["match_id", "venue"]].drop_duplicates(),
    on="match_id"
)
venue_avg_runs = match_runs.groupby("venue")["total_match_runs"].mean()
overall_avg_runs = match_runs["total_match_runs"].mean()
venue_run_factor = venue_avg_runs / overall_avg_runs
data["venue_run_factor"] = data["venue"].map(venue_run_factor)
venue_avg = (
    data
    .groupby(["player", "venue"])["fantasy_points"]
    .mean()
    .reset_index(name="venue_avg_points")
)
data = data.merge(
    venue_avg,
    on=["player", "venue"],
    how="left"
)
```
