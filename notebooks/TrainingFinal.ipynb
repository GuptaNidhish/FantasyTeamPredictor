```python
data['recent_form'] = (data['last3_avg_points']*0.6 + data['last5_avg_points']*0.3 + data['last10_avg_points']*0.1)
```

```python
data['venue_form'] = data['venue_avg_points'] * data['recent_form']
```

```python
data
```

```python
colsToDrop =  ['last5_avg_points',
              'last10_avg_points',
              'bat_pos_per_match',
              'match_month',
              'stage_Elimination Final',
              'stage_Eliminator',
              'stage_Final',
              'stage_Qualifier 1',
              'stage_Qualifier 2',
              'stage_Semi Final',
              'stage_Unknown']
data = data.drop(columns = colsToDrop)
```

```python
data
```

```python
data['season'] = data['season'].str[:4].astype(int)
data['season'].unique()
```

```python
train = data[data['season'] <= 2022].copy()
test  = data[data['season'] > 2022].copy()
X_train = train.drop(columns=['fantasy_points','player','venue','season','player_role_batsman','player_role_bowler','match_id','player_match_number'])
y_train = train['fantasy_points']

X_test = test.drop(columns=['fantasy_points','player','venue','season','player_role_batsman','player_role_bowler','match_id','player_match_number'])
y_test = test['fantasy_points']

print(X_test.shape)
print(X_train.shape)
print(test.shape)
print(train.shape)
```

```python
from lightgbm import LGBMRegressor
best_params = {'subsample': 0.9, 'num_leaves': 127, 'min_child_samples': 5, 'max_depth': 10, 'learning_rate': 0.02, 'colsample_bytree': 0.6}

final_model = LGBMRegressor(
    n_estimators=600,
    random_state=42,
    reg_alpha = 0.1,
    reg_lambda = 1,
    **best_params
)

final_model.fit(
    X_train,
    y_train,
)
```

```python
y_pred = final_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test, y_pred))
```

```python
test['predictions'] = y_pred
```

```python
import joblib
le_player = joblib.load("/Users/nidhishgupta/Desktop/Dream11_Fantasy Team_Predictor/encoders/player_encoder.pkl")
le_venue = joblib.load("/Users/nidhishgupta/Desktop/Dream11_Fantasy Team_Predictor/encoders/venue_encoder.pkl")
test['player'] = le_player.inverse_transform(test['player'])
test['venue'] = le_venue.inverse_transform(test['venue'])
test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("RMSE:", rmse)
print("R2:", r2)
```

```python
from scipy.stats import spearmanr

corr, _ = spearmanr(y_test, y_pred)
print("Spearman Rank Correlation:", corr)
```

```python
results = X_test.copy()

results['actual_points'] = y_test
results['predicted_points'] = y_pred
results['player'] = test['player']
results['match_id'] = test['match_id']
```

```python
results
```

```python
correct_counts = []

for match in results['match_id'].unique():

    match_df = results[results['match_id'] == match]

    predicted_top11 = match_df.sort_values(
        'predicted_points', ascending=False
    ).head(11)['player']

    actual_top11 = match_df.sort_values(
        'actual_points', ascending=False
    ).head(11)['player']

    overlap = len(set(predicted_top11) & set(actual_top11))

    correct_counts.append(overlap)

print("Average correct players in top 11:", sum(correct_counts)/len(correct_counts))
```
