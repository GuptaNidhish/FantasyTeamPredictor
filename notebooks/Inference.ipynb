```python
historical_data = pd.read_csv("/Users/nidhishgupta/Desktop/Dream11_Fantasy Team_Predictor/data/processed/full_req_data.csv")
```

```python
import joblib
le_player = joblib.load("/Users/nidhishgupta/Desktop/Dream11_Fantasy Team_Predictor/encoders/player_encoder.pkl")
le_venue = joblib.load("/Users/nidhishgupta/Desktop/Dream11_Fantasy Team_Predictor/encoders/venue_encoder.pkl")
historical_data['player'] = le_player.inverse_transform(historical_data['player'])
historical_data['venue'] = le_venue.inverse_transform(historical_data['venue'])
```

```python
historical_data
```

```python
from difflib import get_close_matches
unique_player_list_from_df = historical_data['player'].unique()
unique_players = list(unique_player_list_from_df)
print(unique_player_list_from_df)

def map_player_name(name):
    
    match = get_close_matches(name, unique_players, n=1, cutoff=0.5)
    
    if match:
        return match[0]
    else:
        return None
```

```python
mapped_players = [map_player_name(p) for p in squad]
mapped_players = [p for p in mapped_players if p is not None]
print(mapped_players)
```

```python
unique_venue_list_from_df = historical_data['venue'].unique()
unique_venues = list(unique_venue_list_from_df)
print(unique_venues)
def map_venue_name(name):
    
    match = get_close_matches(name, unique_venues, n=1, cutoff=0.5)
    
    if match:
        return match[0]
    else:
        return None
```

```python
mapped_venues = map_venue_name(venue)
mapped_venues
```

```python
players_df = historical_data[
    historical_data['player'].isin(mapped_players)
].copy()
```

```python
players_df
```

```python
players_df = (
    players_df
    .sort_values('player_match_number')
    .groupby('player')
    .tail(1)
)
```

```python
players_df
```

```python
players_df.columns
X_pred = players_df.drop(columns=['fantasy_points','player','venue','player_match_number','match_id','season','player_role_batsman','player_role_bowler'])

players_df['predicted_points'] = model.predict(X_pred)
```
