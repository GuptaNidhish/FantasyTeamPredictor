```python
from difflib import get_close_matches
unique_player_list_from_df = historical_data['player'].unique()
unique_players = list(unique_player_list_from_df)
print(unique_player_list_from_df)

def map_player_name(name):
    
    match = get_close_matches(name, unique_players, n=1, cutoff=0.5)
    
    if match:
        return match[0]
    else:
        return None
```

```python
mapped_players = [map_player_name(p) for p in squad]
mapped_players = [p for p in mapped_players if p is not None]
print(mapped_players)
```

```python
unique_venue_list_from_df = historical_data['venue'].unique()
unique_venues = list(unique_venue_list_from_df)
print(unique_venues)
def map_venue_name(name):
    
    match = get_close_matches(name, unique_venues, n=1, cutoff=0.5)
    
    if match:
        return match[0]
    else:
        return None
print(historical_data['venue'].value_counts())
```

```python
mapped_venues = map_venue_name(venue)
mapped_venues
```

```python
players_df = historical_data[
    historical_data['player'].isin(mapped_players)
].copy()
```

```python
players_df
```

```python
players_df = (
    players_df
    .sort_values('player_match_number')
    .groupby('player')
    .tail(1)
)
```

```python
players_df['venue'] = venue
players_df
```

```python
players_df = players_df.drop(columns = ['venue_avg_points'])
venue_data = historical_data[historical_data['venue'] == venue][['player','venue_avg_points']].drop_duplicates()
venue_data
```
